```python
import pandas as pd

url = 'https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv'
data = pd.read_csv(url)

x = data['YearsExperience']
y = data['Salary']
```

```python
import matplotlib.pyplot as plt
import matplotlib as mlp

plt.scatter(x, y, marker='x', color='red')
plt.title('Experience/Salary')
plt.ylabel('Salary')
plt.xlabel('Experience')
```

```python
w=0
b=0

def plot_pred(w, b):
    y_pred = x*w + b
    plt.plot(x, y_pred, label='predict')
    plt.scatter(x, y, marker='x', color='red', label='real_data')
    plt.title('Experience/Salary')
    plt.xlabel('Experience')
    plt.xlim(0,12)
    plt.ylabel('Salary')
    plt.ylim(-60,140)
    plt.legend()
    
plot_pred(10,60)
```

```python
#interact 是其中的一個函數，用於簡單地將 Python 函數與控件綁定，實現互動式輸入和輸出。
from ipywidgets import interact

interact(plot_pred, w=(-100,100,1), b=(-100,100,1))
```

```python
@interact(color=['red', 'green', 'blue'])
def show_color(color):
    print(f"Color: {color}")
```

```python
@interact(show_text=True)
def show_message(show_text):
    if show_text:
        print("Hello, World!")
    else:
        print(" ")
```
